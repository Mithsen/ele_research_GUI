```python
import cv2
from darkflow.net.build import TFNet
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import os
import sys
%config InlineBackend.figure_format = 'svg'
```

```python
# define the model options and run

options = {
    'model': 'cfg/yolo2-obj.cfg',
    'load': 'weights/yolo2-obj_7000.weights',
    'threshold': 0.3,
}

tfnet = TFNet(options)
```

```python
ele_name_list2  = ['062']
#ele_name_list2 = ['m182','m188','m191','m194','m196','m199','nat','nay','ramani','rani','wrinckle']
for i in range(0,len(ele_name_list2)):
    name = ele_name_list2[i]
    #test_path =  '/home/mithsen/Pictures/Elephants_images/2.2_add_2.0_and_2.1/'+name
    test_path =  '/home/mithsen/Pictures/Elephants_images/3elephant_face_crop/'+name
    total = len(os.listdir(test_path))
    count = 0
    
    if not os.path.exists('/home/mithsen/Pictures/Elephants_images/automated_ear/'+name):
        os.makedirs('/home/mithsen/Pictures/Elephants_images/automated_ear/'+name)
        os.makedirs('/home/mithsen/Pictures/Elephants_images/automated_ear/'+name+'/wrong')
    else:
        print(name+" folder exist")
    for file_name in os.listdir(test_path):
        count = count + 1 
        img_path = test_path+'/'+file_name
        print('/'+name+"--->"+" "+str(count)+" out of "+str(total), end='')
        img = cv2.imread(img_path)
        
        result = tfnet.return_predict(img)
        if(len(result)==0):
            cv2.imwrite('/home/mithsen/Pictures/Elephants_images/automated_ear/'+name+'/wrong/'+file_name,np.asarray(img))
            print("Not Classified")
        for i in range(len(result)):
            if(i>0):
                print("**************")
                plt.imshow(img)
                plt.show()
            if(result[i]['label']=="ear"):
                x1 = result[i]['topleft']['x'] 
                y1 = result[i]['topleft']['y']
                x2 = result[i]['bottomright']['x']
                y2 = result[i]['bottomright']['y']
                crop_img = img[y1:y2, x1:x2]
                #b,g,r = cv2.split(crop_img)       # get b,g,r
                #rgb_img = cv2.merge([r,g,b])
                #plt.imshow(rgb_img)

                cv2.imwrite('/home/mithsen/Pictures/Elephants_images/automated_ear/'+name+'/'+file_name,np.asarray(crop_img))
                print("OK",end='')
            



    
```
